```python
import numpy as np
import torch
import torch.nn.functional as F
from src.models import GATGRU
from src.data import PodDataset
import matplotlib.pyplot as plt
```

```python
dataset = PodDataset("../data", split="test")
model = GATGRU(in_features=2, hidden_channels=5, num_nodes=11, window_len=60)
model.load_state_dict(torch.load("../models/gat-gru.pt"))
model.eval()
```

```python
test_loss = 0.0
true_time_series = []
pred_time_series = []

with torch.no_grad():
    for i in range(len(dataset)):
        (X, Ax), (y, Ay) = dataset[i]
        p = model(X, Ax)
        test_loss += F.mse_loss(p, y.unsqueeze(0)).item()
        
        true_time_series.append(y.numpy())
        pred_time_series.append(p[0].numpy())

print(f"Test loss {test_loss/len(dataset):.4f}")
```

```python
true_resources = np.stack(true_time_series)
pred_resources = np.stack(pred_time_series)
```

```python
plt.plot(true_resources[:, -1, 0])
plt.plot(pred_resources[:, -1, 1])
```
